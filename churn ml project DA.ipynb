## -----------------Business Understanding-----------------

       The telecommunications sector has become one of the main industries in developed countries. The technical progress and the increasing number of operators raised the level of competition. Companies are working hard to survive in this competitive market depending on multiple strategies. Three main strategies have been proposed to generate more revenues: 

    1. Acquire new customers :Trying to make new Customers
    2. Upsell the existing customers :Upselling is a sales strategy used to convince 
    customers to buy more products — usually at a higher price point. 
    3. Increase the retention period of customers : The goal of customer retention is to maintain the company's current customer base and increase loyalty by providing better service or more benefits. 


            However, comparing these strategies taking the value of return on investment (RoI) of each into account has shown that the third strategy is the most profitable strategy, proving that retaining an existing customer costs much lower than acquiring a new one, in addition to being considered much easier than the upselling strategy. To apply the third strategy, companies have to decrease the potential of customer’s churn, known as "the customer movement from one provider to another".

    In a right-skewed distribution, the mean is usually greater than the median, because the long tail on the right side pulls the mean towards the larger values. The mode, which is the most frequently occurring value, may also be less than the median.

    Some examples of right-skewed distributions are the distribution of income, stock prices, and home values. In these cases, there are a small number of individuals or companies with very high values, while the majority of the population has lower values.

    To visualize a right-skewed distribution, a histogram or a box plot can be used. In a histogram, the bars will be taller on the left side and shorter on the right side, while in a box plot, the box will be shorter on the right side and the whisker will be longer on the same side.

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=5)
missing_column = data["TotalCharges"]
# Reshape the data for input into the imputer
#Next, we assume that df is the DataFrame containing the column with missing values called "Total Charges". 
#We select the "Total Charges" column and reshape the data into a 2D array using the reshape() method. 
#This is required because the KNNImputer expects a 2D array as input.
reshaped_data = missing_column.values.reshape(-1, 1)
# Fit the imputer on the reshaped data
imputer.fit(reshaped_data)
# Transform the data with the fitted imputer
imputed_data = imputer.transform(reshaped_data)

# Replace the missing values with the imputed values in the DataFrame
data["TotalCharges"] = imputed_data

# Verify that there are no missing values in the "Total Charges" column
print(data.isnull().sum())
```

```python
from sklearn.metrics import accuracy_score,recall_score,f1_score,confusion_matrix,precision_score,classification_report
def metric(real,predict):
    acc=accuracy_score(real,predict)
    rec=recall_score(real,predict,pos_label="Yes")
    f1=f1_score(real,predict,pos_label="Yes")
    #conmax=confusion_matrix(real,predict)
    pre=precision_score(real,predict,pos_label="Yes")
    clas=classification_report(real,predict)
    print("Accuracy =",acc)
    print("Recall =",rec)
    print("Precision =",pre)
    print("F-score =",f1)
    print(clas)
```
